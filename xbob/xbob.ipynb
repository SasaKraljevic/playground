```python
from pandas import *
pandas.set_option('display.max_rows', 2000) # ovaj code stavljam samo zbog prikaza svih redova tablice
```

```python
# ucitavanje csv datoteke u varijablu xbobTablica i postavljanje naslova kolumni
xbobTablica = read_csv('XBOB-2016-05-12 20-07-41.csv', names=['Rb', 'Artikl', 'Debljina', 'Duzina', 'Sirina'])

# da vidimo kako sada izgleda nasa tablica...
xbobTablica
```

```python
# preracunavanje jedinicnih mjera i dodjela svake kolumne odredjenoj varijabli
xbobTablica['Debljina'] = (xbobTablica['Debljina'] / 1000)

xbobTablica['Duzina'] = (xbobTablica['Duzina'] / 1000000).round(2)
kolumnaDuzina = xbobTablica['Duzina']

xbobTablica['Sirina'] = (xbobTablica['Sirina'] / 1000000).round(2)
kolumnaSirina = xbobTablica['Sirina']

# sada tablici dodajemo novu kolumnu - Povrsina
kolumnaPovrsina = kolumnaDuzina * kolumnaSirina
xbobTablica['Povrsina'] = kolumnaPovrsina.round(2)
```

```python
# sada je nasa tablica malo izmijenjena
xbobTablica
```

```python
# grupiranje tablice po Artiklu i Debljini, a zatim zbrojimo vrijednosti u kolumnama Duzina, Sirina i Povrsina
# da su samo iverali koristio bih .agg({'Povrsina': sum}) kako bih vidio samo ukupnu povrsinu za odredjeni artikl
# ja sam stavio i Duzinu i Sirinu jer nas za RP zanima ukupna Duzina, a Sirinu sam ostavio bez nekog razloga
xbobTablica = xbobTablica.groupby(['Artikl', 'Debljina']).agg({'Duzina': sum,
                                                               'Sirina': sum,
                                                               'Povrsina': sum})
# a ovako izgleda kada imamo sve podatke koji nas zanimaju, u svega nekoliko koraka :)
xbobTablica
```
