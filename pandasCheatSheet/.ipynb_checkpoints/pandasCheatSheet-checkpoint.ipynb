```python
# otvaranje dodatne datoteke
london = read_csv('london_2014.csv', skipinitialspace=True)
london.columns
```

```python
london.head()
```

```python
# prikaz odredjene kolumne
london[[5]].head()
```

```python
london[['WindDirDegrees<br>']].head()
```

```python
# preimenovanje naziva kolumne
london = london.rename(columns={'WindDirDegrees<br>' : 'WindDirDegrees'})
london[['WindDirDegrees']].head()
```

```python
# brisanje nepotrebnih znakova ( <br> )
london['WindDirDegrees'] = london['WindDirDegrees'].str.rstrip('<br>')
london[['WindDirDegrees']].head()
```

```python
# da li postoji NaN u odredjenoj kolumni??
london[london['Events'].isnull()]
```

```python
# mijenjanje vrijednosti NaN
london['Events'] = london['Events'].fillna('')
london[london['Events'].isnull()]
```

```python
# provjera vrste podataka u svim kolumnama
london.dtypes
```

```python
# promjena vrste podataka u kolumnama
london['WindDirDegrees'] = london['WindDirDegrees'].astype('float64')
london['WindDirDegrees']
```

```python
# prikazi kolumne u kojima je smijer vjetra veci od 350 stupnjeva
london[london['WindDirDegrees'] > 350]
```

```python
# brisanje posljednjeg reda
london = london[0:-1]
```

```python
london['GMT']
```

```python
london['GMT'] = to_datetime(london['GMT'])
```

```python
# prikazi samo red sa odredjenim datumom
london[london['GMT'] == datetime(2014, 6, 4)]
```

```python
# prikazi samo redove sa odredjenim smjerom vjetra
london[london['WindDirDegrees'] == 30]
```

```python
# prikazi samo redove sa odredjenim datumima
london[(london['GMT'] >= datetime(2014, 12, 8))
     & (london['GMT'] <= datetime(2014, 12, 12))]
```

```python
# prikazi redove gdje je smjer vjetra veci ili jednak 350 ili manji ili jednak 10
london[(london['WindDirDegrees'] >= 350)
     | (london['WindDirDegrees'] <= 10)]
```

```python
# get and display all the rows in the dataframe that are beween 1 April 2014 and 11 April 2014.
london[(london['GMT'] >= datetime(2014, 4, 1))
     & (london['GMT'] <= datetime(2014, 4, 11))]
```

```python
# Max Gust SpeedKm/h
london[london['Max Gust SpeedKm/h'].isnull()]
```

```python
london['Max Gust SpeedKm/h'] = london['Max Gust SpeedKm/h'].fillna('')
london[london['Max Gust SpeedKm/h'].isnull()]
```
